```python
import pandas as pd
import numpy as np

IPLMatches = pd.read_csv("/kaggle/input/ipl/matches.csv")
ddf = pd.read_csv("/kaggle/input/ipl/deliveries.csv")
print(ddf.shape)
ddf.head()
```

```python
# total bowler
len(ddf['bowler'].value_counts())
```

```python
tempbowler = ddf['bowler'].value_counts().index
fullMatches = {}
for bowler in tempbowler:
    fullMatches[bowler] = len(ddf['match_id'].loc[ddf['bowler'] == bowler].unique())
#fullMatches
```

```python
len(fullMatches)
```

```python
bowler = []
for k,v in fullMatches.items():
    bowler.append(k)
len(bowler)
```

## Economy

```python
runscored = {}
ballsbowled = {}
for b in bowler:
    runscored[b] = sum(ddf['total_runs'].loc[ddf['bowler'] == b ])
    ballsbowled[b] = len(ddf['total_runs'].loc[ddf['bowler'] == b ])
print(runscored)
print(ballsbowled)
```

```python
def economy(runs,balls):
    return (runs * 6 / balls)
```

```python
BowlerEconomy = {}
for b in bowler:
    BowlerEconomy[b] = economy(runscored[b], ballsbowled[b])
BowlerEconomy
```

## Wicket Taking Ability

```python
wicketsTaken = {}
for b in bowler:
    wicketsTaken[b] = len(np.where((ddf['player_dismissed'].loc[ddf['bowler'] == b ]).isnull() == False)[0])
print(wicketsTaken)
```

```python
zerowick = []
for k, v in list(wicketsTaken.items()):
    if v == 0:
        zerowick.append(k)
        #del wicketsTaken[k]
```

```python
def wicktakabil(balls,wickets):
    return (balls/wickets)
    
    
```

```python
WicketTakingAbility = {}
for b in bowler:
    if b in zerowick:
        WicketTakingAbility[b] = 0
    else:
        WicketTakingAbility[b] = wicktakabil(ballsbowled[b],wicketsTaken[b])
WicketTakingAbility
```

## Consistency

```python
def consistency(runs,wickets):
    return (runs/wickets)
```

```python
Consistency = {}
for b in bowler:
    if b in zerowick:
        Consistency[b] = 0
    else:
        Consistency[b] = consistency(runscored[b],wicketsTaken[b])

Consistency
```

```python
colnames = ['Bowler','Economy', 'Wicket Taking Ability', 'Consistency']
BowlerData = pd.DataFrame(columns=colnames)
BowlerData['Bowler'] = bowler
BowlerData['Economy'] = BowlerEconomy.values()
BowlerData['Wicket Taking Ability'] = WicketTakingAbility.values()
BowlerData['Consistency'] = Consistency.values()
```
